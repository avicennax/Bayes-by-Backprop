```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
N = 30
noise_std = 2
x_spread = 5

x = np.random.randn(N) * x_spread
w, b = np.random.randn(2) * noise_std
y = w*x + b + np.random.randn(N) * noise_std
```

```python
plt.figure(figsize=(8,6))
plt.plot(x, y, 'ko');
plt.xlim(-20, 20);
plt.ylim(-20, 20);
```

```python
# Data placeholders
x_ = tf.placeholder(tf.float32, shape=[None], name='x')
y_ = tf.placeholder(tf.float32, shape=[None], name='y')
```

```python
# Variables
with tf.variable_scope("full_model") as full_model:
    w_ = tf.get_variable('w', shape=[])
    b_ = tf.get_variable('b', shape=[])
```

```python
# Loss operation
predictions = x_ * w_ + b_
loss = tf.losses.mean_squared_error(y_, predictions)
```

```python
train_op = tf.train.GradientDescentOptimizer(learning_rate=0.01).minimize(loss)
```

## Training

```python
indices = np.arange(N)
total_epochs = 100
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for epoch in range(total_epochs):
        samples = np.random.choice(indices, 16)
        _, train_loss = sess.run([train_op, loss], feed_dict={x_: x[samples], y_: y[samples]})
        if epoch % int(total_epochs/10) == 0:
            print("loss - {i}/{n}: {loss:.3f}".format(loss=train_loss, i=epoch, n=total_epochs))
        
    w_pred, b_pred = sess.run([w_, b_])
```

```python
plt.figure(figsize=(8,6))
x_bound = np.array([np.min(x) - np.std(x), np.max(x) + np.std(x)])
plt.plot(x, y, 'ko');
plt.plot(x_bound, w_pred*x_bound + b_pred)
plt.xlim(-20, 20);
plt.ylim(-20, 20);
```
